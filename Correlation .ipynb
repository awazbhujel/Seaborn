```python
x=df.corr()
sns.heatmap(x,annot=True)
```

```python
df.corr().unstack().sort_values().drop_duplicates().plot(kind="barh")
```

Note: the one downside is if you have 1.0 correlations that are not one variable to itself, the drop_duplicates() addition would remove them

```python
#Created, Filtered and Charted
dfCorr = df.corr()
filteredDf = dfCorr[((dfCorr >= .5) | (dfCorr <= -.5)) & (dfCorr !=1.000)]
plt.figure(figsize=(30,10))
sns.heatmap(filteredDf, annot=True, cmap="Reds")
plt.show()
```

```python
corr = df.corr() # df is the pandas dataframe
corr.abs().unstack().sort_values(ascending=False)
```

```python
corr = df.corr()
kot = corr[corr>=.1]
plt.figure(figsize=(12,8))
sns.heatmap(kot, cmap="Greens",annot=True)
```
